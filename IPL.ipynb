```python
#Extracting the records where a team won batting first
batting_first = ipl[ipl['win_by_runs'] != 0]
batting_first
```

```python
batting_first.head()
```

```python
# finding out the no of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs']);
plt.title('Distributin of runs');
plt.xlabel('Runs');
```

```python
plt.figure(figsize=(12,4))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]), color="y");
#plt.title(' ');
plt.xlabel('Runs');
```

```python
# making a pie chart to show winning percentage

plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%');
```
